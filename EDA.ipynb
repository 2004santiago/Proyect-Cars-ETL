Cleaning of Duplicate VIN Values

Since the VIN is a unique value that we will use as an identifier, there should be no duplicate VINs. These rows will be deleted.

```python
df = df.drop_duplicates(subset='VIN', keep='first')

frequency = df['VIN'].value_counts()
print(frequency)
```

Cleaning of Price

It contains object-type values with characters such as "$" and ",". Additionally, there are some 'Not Priced' values that cannot be transformed into integers.

```python
filtrated_rows = df[df['Price'] == 'Not Priced']

result = filtrated_rows.head(3)
print(result)
```

Some of these records with "No Price" are displayed... Rows with these specifications will be deleted as they would be considered outliers (Probably "Not Priced" due to dealership advertisement).

```python
df = df[df['Price'] != 'Not Priced']
```

Since we no longer have Not Price, we can make the changes to int format:

```python
df['Price'] = df['Price'].replace('[\$,]', '', regex=True).astype(int)

#Visualization
print(df['Price'].head())
```

Cleaning Deal type

There are several nulls, let's look at the table carefully and replace the nulls. If we delete them just to delete, we would lose a lot of information.

```python
df['DealType'].fillna('NA', inplace=True)
```

```python
#Visualization
print(df['DealType'].value_counts())
```

Indeed there are no longer nulls, the same amount of non-null data is presented as in the other columns

Used/New

Reviewing Used/New, redundant information was found (the brand was mentioned, which is already in Make), we will leave it as a boolean, Used = True, New = False (New will be everything that is not Used, the brand can be seen in the Make column itself)

```python
def used_new_transformation(value):
    if value == 'Used':
        return True

    else:
        return False

df['Used/New'] = df['Used/New'].apply(used_new_transformation)

#Because of the new format (bool) its better if we change the name of the column
#Now its 'Used', wich makes it easly to understand the bool structure of the variable.
#Used = True, Not used (New) = False
df.rename(columns={'Used/New': 'Used'}, inplace=True)
```

```python
#Visualization
print(df['Used'].value_counts())
```

State

Some invalid states are presented, and some are not ideally formatted (or are taken as a separate state)

```python
#We make a lsit of valid states
validated_states = [
    'AL', 'AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA',
    'HI', 'ID', 'IL', 'IN', 'IA', 'KS', 'KY', 'LA', 'ME', 'MD',
    'MA', 'MI', 'MN', 'MS', 'MO', 'MT', 'NE', 'NV', 'NH', 'NJ',
    'NM', 'NY', 'NC', 'ND', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC',
    'SD', 'TN', 'TX', 'UT', 'VT', 'VA', 'WA', 'WV', 'WI', 'WY'
]

#Replace Michigan => MI and AZ-101 to AZ (Arizona)
df['State'] = df['State'].replace({
    'Michigan': 'MI',
    'AZ-101': 'AZ'
})

#Drop the rows with invalid states
df = df[df['State'].isin(validated_states)]
```
